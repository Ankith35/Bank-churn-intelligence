# Bank Churn Intelligence - Notebook 01: Data Profiling

## Objective
Understanding all the columns before any analysis begins.
Every column documented with data type, null count and data relevance.

```python
import pandas as pd
import numpy as nmp
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load Dataset
df = pd.read_csv('../data/raw/Customer-Churn-Records.csv')

print(f"Dataset shape:{df.shape}")
print(f"Rows:{df.shape[0]:,} | Columns:{df.shape[1]}")
print(f"\nColumns names:")
for col in df.columns:
    print(f" --> {col}")
```

```python
# Data type, null checks, unique values
profile = pd.DataFrame({
    'dtype'  : df.dtypes,
    'null_count' : df.isnull().sum(),
    'null_pct' : (df.isnull().sum() / len(df) * 100).round(2),
    'uninue_values' : df.nunique
 })
print(profile)
```

```python
# Churn rate breakdown across key columns
for col in ['Geography','Gender','Card Type','NumOfProducts','Complain','IsActiveMember']:
    print(f"\n{'='*45}")
    print(f"Churn rate by {col}")
    print(f"{'='*45}")
    result = df.groupby(col)['Exited'].agg(['mean','sum','count'])
result.columns = ['Churn Rate','Churned','Total']
result['Churn Rate'] = (result['Churn Rate']*100).round(1).astype(str) + '%'
print(result)
```

```python
# The headline numbers - what this project is really aboubt
churn_rate        = df['Exited'].mean() * 100
total_churned     = df['Exited'].sum()
balance_lost      = df[df['Exited']==1]['Balance'].sum()
avg_bal_churned   = df[df['Exited']==1]['Balance'].mean()
complain_churn    = df[df['Complain']==1]['Exited'].mean() * 100
no_complain_churn = df[df['Complain']==0]['Exited'].mean() * 100

print("=" * 45)
print("   EXECUTIVE SUMMARY - FIRST LOOK")
print("=" * 45)
print(f"  Total customers         : {len(df):,}")
print(f"  Customers churned       : {total_churned:,}")
print(f"  Overall churn rate      : {churn_rate:.1f}%")
print(f"  Total balance lost      : ${balance_lost:,.0f}")
print(f"  Avg balance per churner : ${avg_bal_churned:,.0f}")
print("=" * 45)
print(f"\nComplaint --> Churn rate    : {complain_churn:.1f}%")
print(f"No Complaint --> Churn rate : {no_complain_churn:.1f}%")
print("\nEvery customer who complained, churned.")
print("The bank retained ZERO complaining customers.")
print("=" * 45)

```

## Key Findings from Data Profiling

**Finding 1:** Overall churn rate is ~20% - 1 in 5 customers left.

**Finding 2:** 100% of customers who filed a complaint, churned.
This is not just a data finding - this is a service failure alarm.
The bank has zero complaint-to-retention success rate.

**Finding 3:** Total balance lost runs into millions of dollars.
This is a revenue problem, not just a satisfaction problem.

**Implication:** Complaint resolution is completely broken.
Intervention must happen BEFORE the complaint is filed - once a customer complains, they are already gone.
